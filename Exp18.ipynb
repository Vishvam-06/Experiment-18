```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [0, 70, 30, 90, 60],
    'Sales': np.random.randint(100,500,5),
    'Profit': np.random.randint(10, 100, 5)
})
print(df)
```

```python
#1. Area Plot
plt.figure()
plt.fill_between(df['Category'], df['Value'], color='green', alpha=0.25)
plt.title("vishvam")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid") #apply seaborn style
plt.figure(figsize=(7, 4))
plt.fill_between(df['Category'], df['Sales'], color='skyblue', alpha=0.5, label='Sales')
plt.fill_between(df['Category'], df['Profit'], color='orange', alpha=0.5, label='Profit')
plt.title("vishvam")
plt.xlabel("Category")
plt.ylabel("Value")
plt.legend()
```

```python
#2 Pie chart
plt.figure()
plt.pie(df['Value'], labels=df['Category'], autopct='%1.2f%%')
plt.title('Vishvam')
plt.show()
```

```python
#3 Donut Chart

plt.figure()

plt.pie(df['Value'], labels=df['Category'], autopct='%1.1f%%')
centre_circle = plt.Circle ((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Vishvam")
plt.show()
```

```python
#4. Boxplot (Outliners Detection)
plt.figure()
sns.boxplot(x=df['Value'])
plt.title('Vishvam')
plt.show()
```

```python
#5. Heatmap (Correlation)
plt.figure()
corr = df[['Sales', 'Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Vishvam')
plt.show
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Select only numeric columns
corr = df[['Value', 'Sales', 'Profit']].corr()
plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True)
plt.title("Vishvam")
plt.show()
```

```python
#bubble plot( 3 variable Chart)
import matplotlib.pyplot as plt
plt.scatter(df['Sales'], df['Profit'], s=df['Value']*10) #Size of Bubbles
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Vishvam')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Sales', y='Profit',
                size='Value', #Bubble Size
                hue='Value',  #Colour
                data=df,
                sizes=(50,300), #Min-Max Size
                palette = 'viridis')

plt.title("Vishvam")
plt.show()
```
